**Создадим функции для рассчёта показателей**

1. Support
2. Confidence
3. Expected confidence
4. Lift - данный показатель можно выразить через 2 другие функции, но для экономии вычислительного времени просто будем использовать полученные результаты и деление.

```python
def support(a, b, df):
    counter = 0
    for i in range(0, len(df)):
        if df[a][i] > 0 and df[b][i] > 0:
            counter += 1
    return counter/len(df)
```

```python
def confidence(a, b, df):
    counter = 0
    total = 0
    for i in range(0, len(df)):
        if df[a][i] > 0:
            total += 1
            if df[b][i] > 0:
                counter += 1
    return counter/total
```

```python
def expconf(a, b, df):
    counter = 0
    for i in range(0, len(df)):
        if df[b][i] > 0:
            counter += 1
    return counter/len(df)
```

**Теперь заполним итоговую таблицу**

```python
for i in range(0, len(pairs)):
    sup = support(pairs[i][0], pairs[i][1], mba)
    conf = confidence(pairs[i][0], pairs[i][1], mba)
    exconf = expconf(pairs[i][0], pairs[i][1], mba)
    lift = (conf / exconf)
    res['support'][i] = sup
    res['confidence'][i] = conf
    res['expected confidence'][i] = exconf
    res['lift'][i] = lift
```

**Для подсчёта первой таблицы был выбран простой способ, но таблица считается долго из-за итерирования по строкам pandas dataframe. Поэтому в этот раз я решил оптимизировать прежние функии и использовать итерирование по спискам.**

```python
def support(a, b, df):
    a = df[a].values.tolist()
    b = df[b].values.tolist()
    counter = 0
    for i in range(0, len(b)):
        if a[i] > 0 and b[i] > 0:
            counter += 1
    return counter/len(b)
```

```python
def confidence(a, b, df):
    a = df[a].values.tolist()
    b = df[b].values.tolist()
    counter = 0
    total = 0
    for i in range(0, len(a)):
        if a[i] > 0:
            total += 1
            if b[i] > 0:
                counter += 1
    if total != 0:
        return counter/total
    else:
        return 0.0
```

```python
def expconf(a, b, df):
    a = df[a].values.tolist()
    b = df[b].values.tolist()
    counter = 0
    for i in range(0, len(b)):
        if b[i] > 0:
            counter += 1
    return counter/len(b)
```

```python
def compute_mba(result, mba, pairs):
    for i in range(0, len(pairs)):
        sup = support(pairs[i][0], pairs[i][1], mba)
        conf = confidence(pairs[i][0], pairs[i][1], mba)
        exconf = expconf(pairs[i][0], pairs[i][1], mba)
        if exconf != 0:
            lift = (conf / exconf)
        else: 
            lift = 0
        result['support'][i] = sup
        result['confidence'][i] = conf
        result['expected confidence'][i] = exconf
        result['lift'][i] = lift
    return result
```

**Теперь посчитаем все таблицы**

```python
res_bills0 = compute_mba(res_bills0, mba0, pairs)
res_bills1 = compute_mba(res_bills1, mba1, pairs)
res_bills2 = compute_mba(res_bills2, mba2, pairs)
res_сlients0 = compute_mba(res_clients0, mba_clients0, pairs)
res_clients1 = compute_mba(res_clients1, mba_clients1, pairs)
res_clients2 = compute_mba(res_clients2, mba_clients2, pairs)
```

*Создадим функцию для перевода значений в проценты*

```python
def percent_mba(res):
    for i in range(0, len(res)):
        res["support"][i] *= 100
        res["confidence"][i] *= 100
        res["expected confidence"][i] *= 100
    res.columns = ['support(%)', 'confidence(%)', 'lift', 'expected confidence(%)']
    return res
```

```python
mba_clients_list = [res_clients0, res_clients1, res_clients2]
mba_bills_list = [res_bills0, res_bills1, res_bills2]
```

```python
for df in mba_clients_list:
    df = percent_mba(df)
for df in mba_bills_list:
    df = percent_mba(df)
```
